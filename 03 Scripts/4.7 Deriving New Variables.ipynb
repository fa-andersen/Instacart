# 2. Creating a Subset

```python
#Create a subset
df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
#Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-Range Product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-Range Product'
    elif row['prices'] > 15:
        return 'High Range Product'
    else: return 'Not Enough Data'
```

# 3. If-Statements with User-Defined Functions

```python
#Apply the function using If-statements w/ user-defined functions

df['price_range'] = df.apply(price_label, axis=1)
```

# 5. If-Statements with For-loops

```python
#create new column "busiest day" using a for-loop
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Combine with df_ords_prods_merge, first create a result column
df_ords_prods_merge['busiest_day'] = result
```

```python
#create new for loop with busiest_days' (0&1)" and slowest_days (3&4)
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
#When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants
#you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average
#orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.
```

```python
#print frequency of the order_hour_of_day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#create new for loop with 'busiest_period_of_day', into (3) groups: Most/Average/Fewest Orders
result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most Orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    result.append("Average Orders")
  else:
    result.append("Fewest Orders")
```

```python
#create new column 'busiest_period_of_day'
df_ords_prods_merge['busiest_period_of_day'] = result
```
