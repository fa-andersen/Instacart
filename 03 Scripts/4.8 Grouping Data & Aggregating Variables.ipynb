```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

# Q6.
The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all theirorders using the following criteria:
-If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
-If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#create group based on mean product price by user id and prices
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
#create spending flag & column using set criteria using loc() function
ords_prods_merge.loc[(ords_prods_merge['mean_product_price'] < 10), 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[(ords_prods_merge['mean_product_price'] >= 10), 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag',]].head(10)
```

# Q7.
In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag

that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The
criteria for the flag should be as follows:
-If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
-If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
-If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#create group based on median 'days_since_prior_order'
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(15)
```

```python
#create order_frequency flag & column using set criteria using loc() function
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 20), 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 10) & (ords_prods_merge['median_days_between_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 10), 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
#list out rows that show nan
nan_rows = ords_prods_merge[ords_prods_merge['order_frequency_flag'] == 'nan']
```

```python
nan_rows
```

```python
#I can see that the NaN rows based on the order_frequency_flag are because they match with customers who have only placed one order. NaN appears because there is NaN in days_since_prior_order and median_days_between_orders. Hoorah!
```
